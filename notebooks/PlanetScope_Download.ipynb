```python
import pandas as pd
import ee
ee.Authenticate()
ee.Initialize()
```

```python
# This notebook uses bounding box coordinates generated by SkySat_Clip_Bbox.ipynb

BASE_DIR = 'base_dir'
merged_labels = pd.read_csv(BASE_DIR + 'df.csv')
merged_labels = merged_labels[~merged_labels['month'].isnull()]

merged_labels['month'] = merged_labels['month'].astype(int)
merged_labels['month'] = merged_labels['month'].astype(str)
merged_labels['year'] = merged_labels['year'].astype(int)
merged_labels['year'] = merged_labels['year'].astype(str)
```

```python
def get_bbox(row):
  '''Create a bounding box of size 'length x length'

  :param var coords: coordinates
  :type coords: list
  :param var lemgth: number of meters of each side of bounding box
  :type length: int
  :rtype: ee.Geometry.Rectangle
  :return: bounding box around given coordinates
  '''

  # Get individual coordinates
  minlon = row['lon_1']
  minlat = row['lat_1']
  maxlat = row['lat_2']
  maxlon = row['lon_2']

  # Create a list to store all of the coordinates
  points = [minlon, minlat, maxlon, maxlat]

  # Create a geometry with list of coordinates
  bbox_geom = ee.Geometry.Rectangle(points)

  return bbox_geom

def get_pscope(bbox, startdate, enddate):

  ps = ee.ImageCollection("projects/planet-nicfi/assets/basemaps/africa").filterBounds(bbox).filterDate(startdate, enddate)

  return ps
```

```python
# Extract imagery for all rows
for idx, row in merged_labels.iterrows():

  # Bounding box around centroid (150x150 px)
  bbox = get_bbox(row)

  # Time variables for time segment wanted in mosaic
  year = row['year']
  start_date = year + '-' + row['month'] + '-01'
  end_date = year + '-' + row['month'] + '-28'

  # Retrieve image collection for time frame
  img = ee.Image(get_pscope(bbox, start_date, end_date).first()).clip(bbox).multiply(0.0001)

  # Export the composite, specifying scale and region
  name_list = re.split(r"\.\s*", str(row['filename']))
  name = name_list[0]
  task_config = {
      'description': name,
      'scale': 4.77,
      'folder': 'folder'
  }
  print(name)
  task = ee.batch.Export.image.toDrive(img,**task_config)
  task.start()
```
