We move straggling images in folders. We remove fake rows from the df. We re-index and rename images in the folder and df

```python
image_folder = os.path.join(FOLDER_PATH, "old/skysat_clip_512_4326_cogs_attempt_2")
print(len(os.listdir(image_folder)))
```

Remove missing images

```python
merged_df = pd.read_csv(os.path.join(FOLDER_PATH, "merged_labels.csv"))
merged_df = merged_df.iloc[:, 1:]
```

```python
merged_df[merged_df["activity"] == 1]
```

```python
IMG_FOLDER = os.path.join(FOLDER_PATH, "merged")
filenames = os.listdir(IMG_FOLDER)
```

```python
filtered_df = merged_df[merged_df["filename"].isin(filenames)]
```

```python
len(filtered_df)
```

Remove duplicate rows

```python
columns_to_check = ['lat_1', 'lon_1', 'lat_2', "lon_2", "year", "month", "day"]
```

```python
duplicates = filtered_df[filtered_df.duplicated(subset=columns_to_check)]
```

```python
filtered_df.drop(duplicates.index, axis=0,inplace=True)
```

```python
len(filtered_df)
```

Reset Index

```python
filtered_df = filtered_df.reset_index(drop=True)
```

```python
filtered_df
```

Rename images

```python
for index, row in filtered_df.iterrows():
  old_name = os.path.join(IMG_FOLDER, row["filename"])
  new_name = os.path.join(IMG_FOLDER, str(index) + ".tif")
  if old_name != new_name:
    # print(old_name)
    # print(new_name)
    # break
    os.rename(old_name, new_name)
    filtered_df.at[index, "filename"] = str(index) + ".tif"
  if index % 1000 == 0:
    print(index)
```
